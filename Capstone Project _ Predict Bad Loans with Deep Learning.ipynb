So LendingClub may take the result coming out of this tool into consideration, when they are in determination of whether to approve a loan application from a new customer.

Because the data input contains a great number of features in various types, making the input very complex, I decided to skip the conventional regression methods for modeling. Instead, I'll use Neural Networks with multiple hidden layers, A.K.A Deep Learning.

### Comments:
The target variable contains more categories than I expected.

1. It has the 'Current', 'In Grace Period' and several other status which means it's not yet reached the mature date of the loan, so we do not have a "good/bad" label on these loans. 

-- This is OK, because we can simply take these data out of the model training process; instead, they can be used as out-of-time(OOT) validation data. We can score these loans and wait for their mature date to get the real result, so we will get our first batch of ongoing performance results earlier, rather than really have to wait for any new customers to apply.

2. For the rest of the categories, we can reasonably re-group them into Good/Bad buckets. For example:

Good: 'Fully Paid', 'Does not meet the credit policy. Status:Fully Paid'

Bad: 'Charged Off', 'Default', 'Does not meet the credit policy. Status:Charged Off'

-- Either 'In Grace Period' or 'Late' looks like a hint towards the directly of being bad loan. But the bottom line is that, these status are describing the unmatured debt, so we treat them equally as 'Current' and exclude them from analysis. Only use them as OOT validation group. Keep in mind that even if someone may be late for a few installments, he/she still stands a chance to pay off the debt eventually.

I'll assign the Bad bucket as 1, Good as 0, and all the rest as 2 for now to keep data type consistancy for the column, which is fine because this '2' bucket will not go into the model training so it will not interfere.

```python
# Re-group and transform 'loan_status' into numeric column
regroup = {
        'Charged Off': 1,
        'Current': 2,
        'Default': 1,
        'Does not meet the credit policy. Status:Charged Off': 1,
        'Does not meet the credit policy. Status:Fully Paid': 0,
        'Fully Paid': 0,
        'In Grace Period': 2,
        'Late (16-30 days)': 2,
        'Late (31-120 days)': 2
}

df['loan_status'] = df['loan_status'].map(regroup)
```

### Modeling
#### Deep Learning Implementation

```python
# Set aside the 2 for loan_status, leaving only 0 and 1
to_train = df[df['loan_status'].isin([0,1])]
to_pred = df[df['loan_status'] == 2]
```

```python
# Train Test Split
X = to_train.drop('loan_status', axis=1).values
y = to_train['loan_status'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 101)
```

```python
# Normalizing the Data with MinMaxScaler
# To fit and transform the train data; but only transform the test data without fit

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

### Comments:
Since the data has 94 features, we choose to start the first layer of neural networks with 94 nodes. Because the label to predict is binary, so we use 'sigmoid' as the activation function for the final layer.

Check out Wikipedia to see why I pick 'relu' as the activation function for the other layers: https://en.wikipedia.org/wiki/Activation_function

Check out this paper to see why I use 'adam' as optimizer: https://arxiv.org/abs/1412.6980

```python
# Build the model

model = Sequential()

model.add(
        Dense(94, activation='relu')
)

model.add(
        Dense(30, activation='relu')
)

model.add(
        Dense(15, activation='relu')
)


model.add(
        Dense(1, activation='sigmoid')
)

model.compile(
        optimizer='adam', 
        loss='binary_crossentropy', 
        metrics=['accuracy']
)
```

```python
# Start training it
# Use early_stop to prevent too much over fitting

early_stop = EarlyStopping(
        monitor='val_loss', 
        mode='min', 
        verbose=1, 
        patience=25
)

model.fit(
        X_train, 
        y_train, 
        epochs=200, 
        batch_size=256, 
        validation_data=(X_test, y_test),
         callbacks=[early_stop]
)
```

### Model Evaluation and Validation

```python
# Plot the loss
pd.DataFrame(model.history.history)[['loss','val_loss']].plot()
```

```python
# Show the classification report
predictions = model.predict_classes(X_test)

print(
        confusion_matrix(y_test,predictions), 
        '\n', 
        classification_report(y_test,predictions)
)
```

### Comments:
The graph above shows a sign of over fitting since very early stage. The model's overall f1-score for accuracy is 0.79, which is not too bad; but we can see we have too many Type 2 errors (6338) in the prediction.

### Model Refinement
Two ways of refining the model we will try:
1. Add Dropout layers to bring down the overfitting
2. Lower the cut-off line in binary prediction to reduce the Type 2 error, at the cost of increasing Type 1 error. In LendingClub case, Type 2 error is the more serious problem because it devastates its balance sheet; while Type 1 error is not very big deal.

```python
# Refine the model

model_new = Sequential()

model_new.add(
        Dense(94, activation='relu')
)

model_new.add(Dropout(0.2))

model_new.add(
        Dense(30, activation='relu')
)

model_new.add(Dropout(0.2))

model_new.add(
        Dense(15, activation='relu')
)

model_new.add(Dropout(0.2))

model_new.add(
        Dense(1, activation='sigmoid')
)

model_new.compile(
        optimizer='adam', 
        loss='binary_crossentropy', 
        metrics=['binary_accuracy']
)


model_new.fit(
        X_train, 
        y_train, 
        epochs=200, 
        batch_size=256, 
        validation_data=(X_test, y_test),
         callbacks=[early_stop]
)
```

```python
# Plot the loss
pd.DataFrame(model_new.history.history)[['loss','val_loss']].plot()
```

```python
# Change the cut-off line to decrease Type 2 error

predictions_new = (model_new.predict_proba(X_test) >= 0.2).astype('int')

print(
        confusion_matrix(y_test,predictions_new), 
        '\n', 
        classification_report(y_test,predictions_new)
)
```

### Comments:
1. The graph shows that, by adding in Dropout layers, we have reduced the overfitting issue compared with the old model
2. By changing the cut-off line to 0.2 (default is 0.5), we have dramatically brought down the Type 2 error. Even though the overall accuracy f1-score goes down because of this practice, but it's totally contingent upon the company's risk appetite to whether focus more on Type 2 error or Type 1 error.

```python
# Save the model and scaler
dump(scaler, open('scaler.pkl', 'wb'))
model_new.save('my_model_lending_club.h5')
```
